Read (with naked eyes, not computer) the dictionary and find out the definitions of the following 9 columns and copy the text next to the column name
- UNITID: The unique identification number assigned to postsecondary institutions as surveyed through IPEDS.          
- INSTNM: The institutionâ€™s name         
- CITY: The institution's location   (city)            
- ZIP: The institution's location    (zip code)
- ADM_RATE: The admissions rate at each campus     
- UGDS: Includes the number of degree/certificate-seeking undergraduates enrolled in the fall, as reported in the IPEDS Fall Enrollment component.             
- TUITIONFEE_IN: The cost data include the tuition and required fees of the institution. They are provided for in-state students         
- MN_EARN_WNE_P10: Mean earnings for the institutional aggregate of all federally aided students who enroll in an institution each year and who are employed but not enrolled. Earnings are defined as the sum of wages and deferred compensation from all non-duplicate W-2 forms received for each individual, plus positive self-employment earnings from Schedule SE. 

```python
# reimport the data with above 9 columns (ignore all other variables)
df = pd.read_csv('MERGED2017_18_PP.csv',skipinitialspace = True, usecols = ['UNITID','INSTNM','CITY','STABBR','ZIP','ADM_RATE','UGDS','TUITIONFEE_IN','MN_EARN_WNE_P10'])
```

```python
# Find out the 10 colleges that are hardest to get in (top 10)
mydf = df[df.get('ADM_RATE') != 0]
print(mydf.nsmallest(10,'ADM_RATE'))
```

```python
# 10. Find out the average in-state tuition of top 10
mycol = df.get('TUITIONFEE_IN')
my10 = mycol.nlargest(10)
print("Average in state tuition for the top 10 reporting schools: $", my10.mean())
```

```python
# Find out the average earnings of graduates from the top 10
df5 = df.nlargest(10,'ADM_RATE')
mycol3 = df5.get('MN_EARN_WNE_P10')
print("The average earnings of graduates from the bottom 10 schools is: $", mycol3.mean(),"because data is listed as NULL for the 2017-2018 year")
```
